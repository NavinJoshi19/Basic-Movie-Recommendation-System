```python
#Import Pandas library
```

```python
import pandas as pd
```

```python
#Read both the csv files
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv') 
```

```python
ratings.head()
```

```python
movies.head()
```

```python
#Merge both the files
```

```python
ratings = pd.merge(movies,ratings)
```

```python
ratings.head()
```

```python
#Drop columns which are not useful to us
```

```python
ratings.drop(['genres','timestamp'], axis = 1)
```

```python
user_ratings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
```

```python
user_ratings.head()
```

```python
#Removing movies which less than 10 users have rated and replace Nan with 0
```

```python
user_ratings = user_ratings.dropna(thresh = 10,axis =1)
```

```python
user_ratings = user_ratings.fillna(0)
```

```python
user_ratings.head()
```

```python
#Building similarity matrix
```

```python
similarity = user_ratings.corr(method = 'pearson')
```

```python
similarity.head(20)
```

```python
#Function to get similar movies
```

```python
def get_similar_movies(movie_name,user_rating):
    similar_score = similarity[movie_name] * (user_rating -2.5)
    similar_score = similar_score.sort_values(ascending = False)
    
    return similar_score
```

```python
romantic_movies_user = [("(500) Days of Summer (2009)",5),("Alice in Wonderland (2010)",3),("Aliens (1986)",1),("2001: A Space Odyssey (1968)",2)]

similar_movies = pd.DataFrame()

for movie,rating in romantic_movies_user:
    similar_movies = similar_movies.append(get_similar_movies(movie,rating),ignore_index = True)

similar_movies.head(10)
```

```python
similar_movies.sum().sort_values(ascending=False).head(20)
```

```python
#Recommendation for comedy movies
```

```python
comedy_movies_user = [("Toy Story (1995)",5),("Father of the Bride Part II (1995)",4),("Aliens (1986)",1),("2001: A Space Odyssey (1968)",2)]

similar_movies = pd.DataFrame()

for movie,rating in comedy_movies_user:
    similar_movies = similar_movies.append(get_similar_movies(movie,rating),ignore_index = True)

similar_movies.head(10)

similar_movies.sum().sort_values(ascending=False).head(20)
```
